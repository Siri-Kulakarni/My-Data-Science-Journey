# Data Visualization

Data visualization is an important component of Exploratory Data Analysis (EDA) because it allows a data analyst to “look at” their data and get to know the variables and relationships between them.
Exploratory data analysis (EDA) is a task of analyzing data using simple tools from statistics, simple plotting tools.
Tyeps of Exploratory Data Analysis (Based on number of features we take):
1. Univariate - Analysis involving a single variable. Makes use of histograms, bar charts, pie charts, boxplots. The variable can be categorical or numerical.
2. Bivariate - Analysis involving two varaibles. They can be numerical or categorical. 
- Numerical vs Numerical - Use scatterplot
- Numerical vs Categorical - ANOVA
- Categorical vs Categorical - Chi squared error test
3. Multivariate - Multivariate data arises from more than one variable.

### Subplots:

```python
x = np.linspace(1, 15, 10)
y=x**2
plt.subplot(1, 2, 1)
plt.plot(x, y, 'b--')
plt.subplot(1, 2, 2)
plt.plot(x, y, 'r*')
```

### Piecharts

```python
labels = ["Haley", "Alex", "Luke", "Manny"]
numbers = [50, 90, 100, 55]
fig = plt.figure(figsize = (10, 10))
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #[left, bottom, width, height]
axes.pie(numbers, labels=labels, autopct = "%0.2f%%", colors = ["Pink", "Green", "Red", "Purple" ])
axes.legend() #Adding legend
```

#### Using Explode

```python
Explode = [0.1, 0, 0, 0]
plt.pie(numbers, labels=labels, explode = Explode)
```

## Seaborn

It is built on matplotlib. It has many inbuilt functions. Visualization of univariate and multivariate is easy with Seaborn.

```python
import seaborn as sns
```

There are many datasets in Seaborn by default.

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('tips')
df
```

```python
sns.__version__
```

```python
df.info()
```

### Scatterplot

Now, to see the relation between the tip received by the waiter to the total bill, we can see the scatter plot of this data.

```python
sns.scatterplot(x="total_bill", y="tip", data = df)
```

We can see that, as the bill amount increases, tip also increases.

```python
sns.scatterplot(x = "size", y = "total_bill", data = df)
```

As the number of people per table(size) increases, the tip amount also increases.

### Lineplot

```python
sns.lineplot(x = "size", y = "total_bill", data = df)
```

With increase in number of people, the total bill also increases. In line plot, the semi transparant region shows confidence interval(range). If you do not want the confidence interval, you can use ci = None or errorbar = None.

```python
sns.lineplot(x = "size", y = "total_bill", data = df, ci = None)
```

### Distribution plot

```python
sns.distplot(df["tip"])
```

### Relationship plot

```python
sns.relplot(x = "total_bill", y = "tip", data = df, hue = "day")
```

Hue parameter helps to color plot aspects based on the values of a specific variable

```python
df["day"].value_counts()
```

### Bar plot

```python
sns.barplot(data = df, x = "day", y = "tip")
```

### Joint plot

```python
sns.jointplot(x = "total_bill", y = "tip", data = df)
```

Jointplot is used for both univariate and bivariate analysis. Histograms are representing the univariate analysis and the scatterplot is representing the bivariate analysis.

### Box plot

Boxplots are used for indentifying outliers.

```python
sns.boxplot(x = "day", y = "total_bill", data = df)
```

The dots indicate the outliers. The line which is splitting the block is the median. The lower boundary of the box is the lower quartile and the upper boundary is the upper quartlie.
